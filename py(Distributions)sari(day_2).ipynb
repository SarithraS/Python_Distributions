```python
                                       #BINOMIAL DISTRIBUTION

#binomial distribution --> prob of occ of each trial(0.5)for both head and tail.size of result random.
#A random value can have number of trials


import seaborn as sb
import matplotlib.pyplot as plt
from numpy import random
a=random.binomial(n=10,p=0.5,size=1000)
sb.distplot(a,kde=False)
plt.show()
```

```python
#lam-->the num of occ the event occurs
#POISON DISTRIBUTION

from numpy import random
import seaborn as sb
a=random.poisson(lam=3,size=100)
print(a)
```

```python
#UNIFORM DISTRIBUTION
#equal distribution

#LOGISTIC DISTRIBUTION
#from the graph of given data set it will take decision



from numpy import random
import seaborn as sb
a=random.logistic(loc=1,scale=2,size=10)
print(a)

```

```python
#NORMAL DISTRIBUTION

import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
x=random.normal(loc=10,scale=5,size=500)
sb.distplot(x,hist=False)
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
from numpy import random
a=random.binomial(n=10,p=0.5,size=1000)
sb.distplot(a,kde=False)
plt.show()
```

```python
#POISSON DISTRIBUTION

from numpy import random
import seaborn as sb
import matplotlib.pyplot as plt
a=random.poisson(lam=3,size=1000)
sb.distplot(a,kde=False)
plt.show()
```

```python

#UNIFORM DISTRIBUTION
from numpy import random
import seaborn as sb
import matplotlib.pyplot as plt
a=random.uniform(size=(3,3))
sb.distplot(a,hist=False)
plt.show()
```

```python
#LOGISTIC DISTRIBUTION

from numpy import random
import seaborn as sb
import matplotlib.pyplot as plt
a=random.logistic(loc=1,scale=2,size=1000)
sb.distplot(a,hist=False)
plt.show()
```

```python
#MULTINOMIAL DISTRIBUTION
from numpy import random
import seaborn as sb
import matplotlib.pyplot as plt
a=random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6],size=10)
sb.distplot(a)
plt.show()
```

```python
from numpy import random
# random numbers---->multinomial
a=random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6],size=10)
print(a)
```

```python
#EXPONENTIAL DISTRIBUTION
from numpy import random
import seaborn as sb
import matplotlib.pyplot as plt
a=random.exponential(size=1000)
sb.distplot(a,hist=False)
plt.show()

```

```python
#CHI SQUARE

from numpy import random
import seaborn as sb
import matplotlib.pyplot as plt
a=random.chisquare(df=1,size=1000)
sb.distplot(a,hist=False)
plt.show()
```

```python
#PARETO DISTRIBUTION

from numpy import random
import seaborn as sb
import matplotlib.pyplot as plt
a=random.pareto(a=3,size=100)
sb.distplot(a,kde=False)
plt.show()
```
